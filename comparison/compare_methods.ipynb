```python
import numpy as np
import matplotlib.pyplot as plt

# The four metrics your assignment requires
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

# *** CHANGE THESE NUMBERS! ***
# Put your TextBlob scores here in this exact order: [Accuracy, Precision, Recall, F1-Score]
textblob_scores = [0.55, 0.85, 0.55, 0.64] 

# Put your Random Forest scores here in the same order:
rf_scores = [0.83, 0.81, 0.83, 0.80] 

x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, textblob_scores, width, label='TextBlob', color='skyblue')
bars2 = ax.bar(x + width/2, rf_scores, width, label='Random Forest', color='salmon')

ax.set_ylabel('Performance Score')
ax.set_title('Complete Evaluation: TextBlob vs. Random Forest')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend(loc='lower right')
ax.set_ylim(0, 1.1) # Set a little above 1.0 to leave room for the text

# This adds the exact numbers on top of every single bar
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f"{yval:.2f}", ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
